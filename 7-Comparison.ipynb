```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data = {
    'Toolkit': ['Recall','Precision','F1-Score'],
    'H2O AutoML': [0.8367, 0.9213,0.8770],
    'EvalML': [0.8163, 0.8333,0.8247],
    'PyCaret': [0.7959, 0.9630,0.8715],
    'TPOT': [0.7857, 0.9506,0.8603],
    'Auto-sklearn': [0.8265, 0.9759,0.8950],
    'AutoKeras': [0.7347, 0.8889,0.8045]
}

df = pd.DataFrame(data)

df_melted = df.melt(id_vars=['Toolkit'], var_name='Metric', value_name='Score')

sns.set(style="whitegrid")

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']

plt.figure(figsize=(12, 6))
sns.barplot(x='Toolkit', y='Score', hue='Metric', data=df_melted, palette=colors, edgecolor='black')
plt.title('Metrics Results for Different AutoML Toolkits', fontsize=16)
plt.ylabel('Result', fontsize=14)
plt.xlabel('Metrics', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.legend(title='AutoML Toolkits', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axhline(y=0.8, color='gray', linestyle='--', linewidth=0.5)  # Add horizontal dashed line
plt.tight_layout()
plt.show()
```
